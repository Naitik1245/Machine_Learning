```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Melbourne_housing.csv')
```

```python
df
```

```python
df.nunique()
```

```python
df.shape
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount',
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = df[cols_to_use]
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
cols_to_fill_0=['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
```

```python
df[cols_to_fill_0]=df[cols_to_fill_0].fillna(0)
```

```python
df['Landsize']=df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea']=df['BuildingArea'].fillna(df.Landsize.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

```python
x=df.drop('Price',axis='columns')
y=df['Price']
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

**How To AutoTune Regualarisation Parameters**

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv=LassoCV(alphas=None,cv=5,max_iter=1000)
```

```python
lasso_cv.fit(x_train,y_train)
```

```python
lasso_cv.score(x_train,y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['Propertycount','Landsize','BuildingArea']] = scaler.fit_transform(df[['Propertycount','Landsize','BuildingArea']])

x = df[['Propertycount','Landsize','BuildingArea']] # Select the scaled features
y = df['Price']


from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
lasso_cv=LassoCV(alphas=None,cv=5,max_iter=1000)
```

```python
lasso_cv.fit(x_train,y_train)
```

```python
lasso_cv.score(x_train,y_train)
```

```python
lasso_cv.score(x_test, y_test)
```

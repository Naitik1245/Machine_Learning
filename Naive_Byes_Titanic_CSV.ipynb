```python
import pandas as pd
df=pd.read_csv('titanic.csv')
df
```

```python
df.drop(['PassengerId','Name','SibSp','Cabin','Embarked','Parch','Ticket'],axis='columns',inplace=True)
df
```

```python
target=df.Survived
inputs=df.drop('Survived',axis='columns')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Encoder=LabelEncoder()
inputs['Sex']=le_Encoder.fit_transform(inputs['Sex'])
```

```python
df
```

```python
import math
median_age=math.floor(df.Age.median())
inputs.Age=inputs.Age.fillna(median_age)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.4)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test)
```
